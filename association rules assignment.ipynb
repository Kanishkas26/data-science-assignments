```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import association_rules, apriori
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel(r'C:\Users\KANISHKA\Downloads\Association Rules\Association Rules\Online retail.xlsx',header=None)
```

```python
df
```

```python
df.columns = ['Food Items']
```

```python
df
```

```python
## splillting the data into appropriate format
```

```python
def txt_split(txt):
    return txt.split(',')
```

```python
data = df['Food Items'].apply(txt_split)
```

```python
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
encoded_df = te.fit_transform(data)
```

```python
encoded_df
```

```python
te.columns_
```

```python
data = pd.DataFrame(encoded_df, columns=te.columns_)
```

```python
data
```

```python
data.replace([True,False],[1,0],inplace=True)
```

```python
data
```

```python
scores = apriori(data, min_support = 0.001, use_colnames = True)
```

```python
rules = association_rules(scores)
rules
```

```python
rules.sort_values(by='confidence', ascending=False)
```
